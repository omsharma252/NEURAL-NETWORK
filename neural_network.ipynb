# ─────────────────────────────────────────────────────────────
# Question 2: Explain the basic architecture and functioning
# of a Perceptron. What are its limitations?
# ─────────────────────────────────────────────────────────────

"""
Answer:

A Perceptron is the simplest form of an artificial neural network. It is a single-layer
model that takes multiple inputs, applies weights, sums them up with a bias, and passes
the result through an activation function to produce a binary output.

Architecture of a Perceptron:

Input Layer:
Receives n input features x1, x2, ..., xn representing the data.

Weights:
Each input has a corresponding weight w1, w2, ..., wn that determines its importance.

Bias:
An additional learnable parameter b that shifts the decision boundary.

Weighted Sum (Summation):
z = (w1*x1) + (w2*x2) + ... + (wn*xn) + b

Activation Function:
A step function is applied: output = 1 if z >= 0, else output = 0

Functioning:
The perceptron receives inputs and multiplies each by its weight. It sums all weighted
inputs and adds the bias. This result is passed through the step function to give a
binary output. If the prediction is wrong, weights are updated using the Perceptron
Learning Rule:
    w_new = w_old + (learning_rate * error * input)
    where error = actual_output - predicted_output

This process repeats until all samples are correctly classified or a maximum number of
epochs is reached.

Limitations of Perceptron:

Linear Separability:
A perceptron can only solve linearly separable problems. It completely fails on problems
like XOR where classes cannot be separated by a straight line.

Binary Output Only:
It produces only 0 or 1. It cannot give probability values or handle multi-class output.

No Hidden Layers:
It has no hidden layers, so it cannot learn complex or hierarchical patterns in data.

Non-differentiable Activation:
The step function is not differentiable, which means gradient-based optimization like
backpropagation cannot be applied directly.

Sensitive to Feature Scale:
Performance degrades significantly when input features are on different scales.
"""


# ─────────────────────────────────────────────────────────────
# Question 3: Describe the purpose of activation functions in
# neural networks. Compare Sigmoid, ReLU, and Tanh functions.
# ─────────────────────────────────────────────────────────────

"""
Answer:

Purpose of Activation Functions:
Activation functions introduce non-linearity into neural networks. Without them, no
matter how many layers a network has, the entire model would behave like a single linear
transformation and would be unable to learn complex patterns. Activation functions allow
the network to model non-linear relationships between inputs and outputs. They also
control the range of values passed to the next layer, which helps with training
stability.

Sigmoid Function:
Formula: f(x) = 1 / (1 + e^(-x))
Output Range: (0, 1)
The sigmoid function squashes all inputs to a value between 0 and 1, making it useful
for binary classification output layers. However, it suffers from the vanishing gradient
problem when inputs are very large or very small, since the gradient becomes nearly zero.
It is also not zero-centered, which can slow down training in deep networks.

ReLU (Rectified Linear Unit):
Formula: f(x) = max(0, x)
Output Range: [0, infinity)
ReLU is currently the most widely used activation function in hidden layers. It is
computationally simple, does not suffer from vanishing gradient for positive values, and
produces sparse activations which improves efficiency. The main drawback is the dying
ReLU problem where neurons can get stuck outputting zero for all inputs if weights
become too negative during training.

Tanh (Hyperbolic Tangent):
Formula: f(x) = (e^x - e^(-x)) / (e^x + e^(-x))
Output Range: (-1, 1)
Tanh is similar to sigmoid but is zero-centered, meaning its output ranges from -1 to 1.
This makes it generally better than sigmoid for hidden layers because the gradients are
stronger near the center. However, it still suffers from vanishing gradient at extreme
values. It is commonly used in hidden layers and recurrent neural networks.

Comparison Summary:
Sigmoid is best for output layers in binary classification. ReLU is the default choice
for hidden layers in deep networks. Tanh works well in hidden layers when zero-centered
output is needed, especially in RNNs.
"""


# ─────────────────────────────────────────────────────────────
# Question 4: What is the difference between Loss function and
# Cost function in neural networks? Provide examples.
# ─────────────────────────────────────────────────────────────

"""
Answer:

Loss Function:
A loss function measures the error between the predicted output and the actual output
for a single training example. It quantifies how wrong the model is for one particular
data point.

Example (Binary Cross-Entropy for one sample):
    Loss = -[y * log(y_hat) + (1 - y) * log(1 - y_hat)]
If actual y = 1 and predicted y_hat = 0.8:
    Loss = -[1 * log(0.8) + 0 * log(0.2)] = 0.223

Cost Function:
A cost function is the average of the loss function computed over the entire training
dataset. It gives an overall measure of how well (or poorly) the model is performing
across all training examples. The goal of training is to minimize this cost function.

Example (Mean Squared Error as Cost):
    Cost = (1/N) * sum of (y_i - y_hat_i)^2 for all N training samples

Key Differences:

Scope:
Loss function operates on one training example. Cost function operates on the entire
training dataset.

Purpose:
Loss function measures per-sample error. Cost function measures overall model
performance and is what the optimizer actually minimizes.

Relationship:
Cost function = average (or sum) of loss function over all training samples.

Common Examples:
For regression: Mean Squared Error (MSE), Mean Absolute Error (MAE)
For binary classification: Binary Cross-Entropy
For multi-class classification: Categorical Cross-Entropy
For margin-based models: Hinge Loss
"""


# ─────────────────────────────────────────────────────────────
# Question 5: What is the role of optimizers in neural networks?
# Compare Gradient Descent, Adam, and RMSprop.
# ─────────────────────────────────────────────────────────────

"""
Answer:

Role of Optimizers:
Optimizers are algorithms that update the weights and biases of a neural network in
order to minimize the cost function. During training, the optimizer uses the gradients
computed by backpropagation to determine how much and in which direction to adjust each
weight. Without an optimizer, a neural network has no mechanism to learn from data.

The general weight update rule is:
    w = w - learning_rate * gradient_of_cost_with_respect_to_w

Gradient Descent:
Gradient Descent is the most basic optimizer. It computes the gradient using the entire
training dataset before making a single weight update. This makes it very slow for large
datasets. It uses a fixed learning rate and can get stuck in local minima. A variant
called Stochastic Gradient Descent (SGD) updates weights after each sample, and
Mini-batch Gradient Descent updates after each small batch of samples.

RMSprop (Root Mean Square Propagation):
RMSprop adapts the learning rate for each parameter individually by dividing the
gradient by a running average of its recent squared values. This prevents the learning
rate from growing too large or becoming too small. It works well for non-stationary
problems and recurrent neural networks. The main hyperparameter is rho (decay factor),
typically set to 0.9.

Adam (Adaptive Moment Estimation):
Adam combines the advantages of both momentum and RMSprop. It maintains a moving
average of both the gradient (first moment) and the squared gradient (second moment)
for each parameter. This allows it to adapt the learning rate per parameter while also
benefiting from momentum to accelerate convergence. Adam converges faster than both
Gradient Descent and RMSprop in most cases, handles sparse gradients well, and is
robust to noisy data. Default hyperparameters are learning_rate=0.001, beta1=0.9,
beta2=0.999.

Comparison Summary:
Gradient Descent is simple but slow and not suitable for large datasets. RMSprop is
adaptive and works well for RNNs. Adam is the most commonly used optimizer because it
combines the best features of both methods and works well across most problem types.
"""

```python
# Question 6: Write a Python program to implement a single-layer
# perceptron from scratch using NumPy to solve the logical AND gate.
# ─────────────────────────────────────────────────────────────

import numpy as np

# AND gate inputs and outputs
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([0, 0, 0, 1])

# Initialize weights and bias
weights = np.zeros(2)
bias = 0
learning_rate = 0.1
epochs = 100

def step_function(z):
    return 1 if z >= 0 else 0

# Training the perceptron
for epoch in range(epochs):
    total_error = 0
    for xi, yi in zip(X, y):
        z = np.dot(xi, weights) + bias
        predicted = step_function(z)
        error = yi - predicted
        weights += learning_rate * error * xi
        bias += learning_rate * error
        total_error += abs(error)
    if total_error == 0:
        print(f"Model converged at epoch {epoch + 1}")
        break

print(f"Final Weights : {weights}")
print(f"Final Bias    : {bias}")

print("\nAND Gate - Test Results:")
print(f"{'Input':<15} {'Expected':<12} {'Predicted'}")
print("-" * 38)
for xi, yi in zip(X, y):
    z = np.dot(xi, weights) + bias
    pred = step_function(z)
    print(f"{str(xi):<15} {yi:<12} {pred}")

correct = sum(step_function(np.dot(xi, weights) + bias) == yi for xi, yi in zip(X, y))
print(f"\nAccuracy: {correct}/{len(y)} = {correct/len(y)*100:.1f}%")
```

```python
# Question 8: Use Keras to build and train a simple multilayer
# neural network on the MNIST digits dataset.
# Print the training accuracy.
# ─────────────────────────────────────────────────────────────

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load and preprocess MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

x_train = x_train.reshape(-1, 784).astype("float32") / 255.0
x_test  = x_test.reshape(-1, 784).astype("float32") / 255.0

print(f"Training samples : {x_train.shape[0]}")
print(f"Test samples     : {x_test.shape[0]}")
print(f"Input features   : {x_train.shape[1]}  (28x28 pixels flattened)")

# Build the model
model = keras.Sequential([
    layers.Dense(256, activation='relu', input_shape=(784,)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64,  activation='relu'),
    layers.Dense(10,  activation='softmax')
], name="MNIST_MLP")

model.summary()

# Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train
history = model.fit(x_train, y_train,
                    epochs=10,
                    batch_size=128,
                    validation_split=0.1,
                    verbose=1)

# Evaluate
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)

print(f"\nFinal Training Accuracy : {history.history['accuracy'][-1]*100:.2f}%")
print(f"Final Test Accuracy     : {test_acc*100:.2f}%")
print(f"Test Loss               : {test_loss:.4f}")
```

```python
 Question 9: Visualize the loss and accuracy curves for a neural
# network model trained on the Fashion MNIST dataset.
# Interpret the training behavior.
# ─────────────────────────────────────────────────────────────

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

# Load and preprocess Fashion MNIST
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

x_train = x_train.reshape(-1, 784).astype("float32") / 255.0
x_test  = x_test.reshape(-1, 784).astype("float32") / 255.0

# Build the model
model = keras.Sequential([
    layers.Dense(256, activation='relu', input_shape=(784,)),
    layers.Dropout(0.3),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(64,  activation='relu'),
    layers.Dense(10,  activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train
history = model.fit(x_train, y_train,
                    epochs=15,
                    batch_size=128,
                    validation_split=0.15,
                    verbose=1)

# Plot curves
epochs_range = range(1, len(history.history['loss']) + 1)

plt.figure(figsize=(13, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, history.history['loss'],     label='Training Loss')
plt.plot(epochs_range, history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve - Fashion MNIST')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(epochs_range, history.history['accuracy'],     label='Training Accuracy')
plt.plot(epochs_range, history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curve - Fashion MNIST')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"\nFinal Test Accuracy : {test_acc*100:.2f}%")

"""
Interpretation of Training Behavior:

Loss Curve:
Training loss decreases consistently across epochs, indicating the model is learning
from the data. Validation loss also decreases and follows a similar trend to training
loss, which means the model is generalizing well and not overfitting significantly.
If validation loss had started increasing while training loss kept decreasing, it would
indicate overfitting.

Accuracy Curve:
Training accuracy increases steadily with each epoch. Validation accuracy follows a
similar trend with a small gap below training accuracy. This small gap is expected and
normal. The Dropout layers added helped in keeping this gap small by preventing the
model from memorizing the training data.

Overall the model shows healthy training behavior. Fashion MNIST is a more challenging
dataset than MNIST, so achieving around 88-91% test accuracy is a good result.
"""
```

```python
# Question 10: Real-time Fraud Detection - Deep Learning Workflow
# Design, train, and evaluate a model for bank fraud detection
# on a large, imbalanced dataset.
# ─────────────────────────────────────────────────────────────

"""
Approach and Design Decisions:

Model Design:
A multilayer neural network is the right choice here rather than a single perceptron.
The dataset is large and contains complex feature interactions between transaction
amount, merchant ID, location, and time. A deeper network can capture these patterns
better.

Architecture used:
Input(10 features) -> Dense(128, ReLU) -> Dense(64, ReLU) -> Dense(32, ReLU) -> Dense(1, Sigmoid)

Activation Functions:
ReLU is used in all hidden layers because it is computationally efficient, avoids the
vanishing gradient problem, and works well on tabular/structured data.
Sigmoid is used in the output layer because this is a binary classification problem
(fraud = 1, legitimate = 0) and sigmoid gives a probability between 0 and 1.

Loss Function:
Binary Cross-Entropy is used since we have a binary classification task. To address
class imbalance, class_weight is passed during training so the model penalizes
misclassification of fraud cases more heavily than legitimate ones.

Handling Class Imbalance:
The dataset has very few fraud cases compared to legitimate ones, so accuracy is not a
reliable metric. Instead, Precision, Recall, F1-Score, and AUC-ROC are used. Class
weights are computed as the ratio of majority to minority class and passed to model.fit.

Optimizer:
Adam with learning_rate=0.001 is used because it adapts the learning rate per parameter,
converges faster than standard gradient descent, and handles noisy and imbalanced
gradients effectively.

Preventing Overfitting:
Dropout layers after each hidden layer reduce overfitting by randomly disabling neurons
during training. L2 regularization penalizes large weights. Batch Normalization
stabilizes and speeds up training. EarlyStopping stops training when validation loss
stops improving, restoring the best weights automatically.
"""

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

np.random.seed(42)
tf.random.set_seed(42)

# Simulating an imbalanced fraud detection dataset
n_legit = 9800
n_fraud = 200

X_legit = np.random.randn(n_legit, 10)
X_fraud = np.random.randn(n_fraud, 10) + 2.5

X = np.vstack([X_legit, X_fraud])
y = np.array([0] * n_legit + [1] * n_fraud)

print(f"Total Samples    : {len(y)}")
print(f"Legitimate       : {n_legit}  ({n_legit/len(y)*100:.1f}%)")
print(f"Fraud            : {n_fraud}  ({n_fraud/len(y)*100:.1f}%)")

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Feature scaling
scaler  = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)

# Compute class weights for imbalance
fraud_count = np.sum(y_train == 1)
legit_count = np.sum(y_train == 0)
class_weight = {0: 1.0, 1: legit_count / fraud_count}
print(f"\nClass Weights    : {class_weight}")

# Build the fraud detection model
model = keras.Sequential([
    layers.Dense(128, activation='relu',
                 kernel_regularizer=regularizers.l2(0.001),
                 input_shape=(10,)),
    layers.BatchNormalization(),
    layers.Dropout(0.4),

    layers.Dense(64, activation='relu',
                 kernel_regularizer=regularizers.l2(0.001)),
    layers.BatchNormalization(),
    layers.Dropout(0.3),

    layers.Dense(32, activation='relu'),
    layers.Dropout(0.2),

    layers.Dense(1, activation='sigmoid')
], name="Fraud_Detection_Model")

model.summary()

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy',
             keras.metrics.AUC(name='auc'),
             keras.metrics.Precision(name='precision'),
             keras.metrics.Recall(name='recall')]
)

early_stop = keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=7, restore_best_weights=True
)

# Train
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=256,
    validation_split=0.15,
    class_weight=class_weight,
    callbacks=[early_stop],
    verbose=1
)

# Evaluate
y_pred_prob = model.predict(X_test).flatten()
y_pred      = (y_pred_prob >= 0.5).astype(int)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Legitimate', 'Fraud']))

print(f"ROC-AUC Score : {roc_auc_score(y_test, y_pred_prob):.4f}")

cm = confusion_matrix(y_test, y_pred)
print(f"\nConfusion Matrix:")
print(f"                   Predicted Legit   Predicted Fraud")
print(f"Actual Legit            {cm[0][0]}              {cm[0][1]}")
print(f"Actual Fraud            {cm[1][0]}              {cm[1][1]}")

# Plot training curves
epochs_range = range(1, len(history.history['loss']) + 1)

plt.figure(figsize=(13, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, history.history['loss'],     label='Training Loss')
plt.plot(epochs_range, history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve - Fraud Detection')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(epochs_range, history.history['auc'],     label='Training AUC')
plt.plot(epochs_range, history.history['val_auc'], label='Validation AUC')
plt.title('AUC Score - Fraud Detection')
plt.xlabel('Epochs')
plt.ylabel('AUC')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
